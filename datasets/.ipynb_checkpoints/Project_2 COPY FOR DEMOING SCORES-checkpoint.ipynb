Make dummies variables for all objects in test:

```python
dummies_test = pd.get_dummies(df_test[all_objects_test], drop_first=True)
```

```python
df_test = df_test.drop(['MS Zoning', 'Street', 'Alley', 'Lot Shape', 'Land Contour',
       'Utilities', 'Lot Config', 'Land Slope', 'Neighborhood', 'Condition 1',
       'Condition 2', 'Bldg Type', 'House Style', 'Roof Style', 'Roof Matl',
       'Exterior 1st', 'Exterior 2nd', 'Mas Vnr Type', 'Exter Qual',
       'Exter Cond', 'Foundation', 'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure',
       'BsmtFin Type 1', 'BsmtFin Type 2', 'Heating', 'Heating QC',
       'Central Air', 'Electrical', 'Kitchen Qual', 'Functional',
       'Fireplace Qu', 'Garage Type', 'Garage Finish', 'Garage Qual',
       'Garage Cond', 'Paved Drive', 'Pool QC', 'Fence', 'Misc Feature',
       'Sale Type'], axis = 1)
# Join the dummies dataframe with the remaining data frame
df_test = df_test.join(dummies_test)
df_test.head(5)
```

```python
all_train_column_names = list(df_train.columns.values)
#list of all column names in df_train
```

This creates a list that only contains column names that are in both dataframes. 
This is important because after dummifying we added new columns that are not in both data sets.
We only want to train the data set on columns that are also in the testing data, for higher accuracy:

```python
list_all_same_column_names = []
for i in all_test_column_names:
    for elem in all_train_column_names:
        if i == elem:
            list_all_same_column_names.append(i)
list_all_same_column_names
```

This creates a new dataframe for test that only has columns shared by both dataframes:

```python
df_test = df_test[list_all_same_column_names]
df_test
```

I'm using this so that add back "SalePrice" to my training set:

```python
list_all_same_column_names_with_sale_price =  list_all_same_column_names 
list_all_same_column_names_with_sale_price.append("SalePrice")
```

Creates a new dataframe for train that only has columns shared by both dataframes + "SalePrice":

```python
df_train = df_train[list_all_same_column_names_with_sale_price]
df_train
```

This fills remaining null values of int and float columns with the mean of the column:

This fills remaining null values of int and float columns with the mean of the column:

Create X and y:

```python

X = df_train.drop('SalePrice', axis=1)
y = df_train['SalePrice']
```

Run train test split:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 42)
```

```python
lr = LinearRegression()
```

Fit to my training data:

```python
lr.fit(X_train, y_train)
```

Check r2 scores of training vs. testing data. Good r2 scores that predicts well to unseen data using lr model:

```python
lr.score(X_train, y_train), lr.score(X_test, y_test)
```

Cross val score to see accuracy accross folds. Decent scores (.80's) but the last fold is lower (.698) which suggests improvements may be needed on this model:

```python
cross_val_score(lr, X_train, y_train, cv = 5)
```

0.838 cross val score for lr model shows decent model accuracy (around 83%) across 5 different folds:

```python
cross_val_score(lr, X_train, y_train, cv = 5).mean()
```

Transform to scalar values so I can fit a new model with an apples to apples comparison for all features and then use lasso:

```python
sc = StandardScaler()
Z_train = sc.fit_transform(X_train)
Z_test = sc.transform(X_test)
```

Set up a list of Lasso alphas to check:

```python
#try adjusting tolerance parameter raise from .001
l_alphas = np.logspace(-5, 1, 100)
# Cross-validate over our list of Lasso alphas.
lasso_cv = LassoCV(alphas = l_alphas, cv=5, tol = .01)
# Fit model using best ridge alpha!
lasso_cv.fit(Z_train, y_train)
#cite lecture on Lassos and Ridges
```

Lasso model r2 score is essentially the same as lr model for r2 score. These are good baseline r2 scores that predicts well to unseen data (at around 90% confidence) using lr model:

```python
lasso_cv.score(Z_train, y_train), lasso_cv.score(Z_test, y_test)
```

Cross val score for lasso model aslo has decent scores (.80's) but the last fold is lower (.698):

```python
cross_val_score(lasso_cv, Z_train, y_train, cv = 5)
```

0.838 cross val score for lasso model shows decent model accuracy (around 84%) across 5 different folds:

```python
cross_val_score(lasso_cv, Z_train, y_train, cv = 5).mean()
```

Prints out all x column names:

This is a list of every coeficient in my lasso function paired with its column name:

This is a list of variables that have the smallest scaled coeficients and therefore these columns have the least predictive power in my model. I will reiterate my model by dropping these columns and trying a polynomial fit. I continued to increase my bounds and add more columns until I have dropped enough columns to have a shape of polynomial features that makes sense in the data set. An example of a shape that makes sense is (w, z) where w>z. See this below when I measure the shape of polynomial features.

```python
my_coefficients= list(zip(X.columns, lasso_cv.coef_))
#print([my_coeficients])
my_list_of_deletes = []
for i in range(0, len(my_coefficients)):
    my_coefficients[i][1]
    if my_coefficients[i][1] < 3000 and my_coefficients[i][1] > -3000:
            my_list_of_deletes.append(my_coefficients[i][0])
my_list_of_deletes
```

Set up a new data frame to try making a new model that has fewer features/columns and (hopefully) has more predictive power. I am dropping columns with the lowest predictive power
as shown by smallest absolute value of scaled lasso coeficients

```python
df_train_2 = df_train.drop(my_list_of_deletes, axis=1)
```

Printing out shape to make sure my train 2 and test 2 dats has been manipulated in the same way after dropping columns. This output makes sense because training data has one extra columns, Sale Price.

```python
df_train_2.shape
```

Repeating the above process for dropping featuees, now in my testing data:

```python

df_test_2 = df_test.drop(my_list_of_deletes, axis=1)
```

```python
df_test_2.shape
```

Here I am setting my new X and Y ((X_2), (y_2)) to use in my model with new feature engineering.

```python

X_2= df_train_2.drop('SalePrice', axis=1)
y_2 = df_train_2['SalePrice']
```

Here I am setting my new X (X_2_test_to_submit) to use for my testing data that now has new feature engineering.

```python
X_2_test_to_submit= df_test_2
```

Here I am making sure that the features match for my X_2 training data and my X_2_test_to_submit testing data:

```python
X_2_test_to_submit.shape, X_2.shape
```

Here I am using a polynomial fit on my newly engineered training data. I am calling the shape to make sure that my X_poly has a shape of polynomial features that makes sense in the data set. An example of a shape that makes sense is (w, z) where w>z. See this below when I measure the shape of polynomial features. This is referenced above as a metric used in my feature engineering function from cell 39.

```python
# Instantiate our PolynomialFeatures object to create all two-way terms.
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)

# Fit and transform our X data.
X_poly = poly.fit_transform(X_2)
X_poly.shape
```

```python
Here I am using a polynomial fit on my newly engineered testing data.
```

```python
#fit the poly to my set of original testing data
X_poly_test = poly.fit_transform(X_2_test_to_submit)
X_poly_test.shape
```

Making my train test split on my training data:

```python
# Create train/test splits.

X_2_train, X_2_test, y_2_train, y_2_test = train_test_split(
    X_poly,
    y_2,
    test_size=0.15,
    random_state=42
)
```

Calling a new linear regression to fit to my training data that now has polynomial features:

```python

new_lr = LinearRegression()
```

```python
new_lr.fit(X_2_train, y_2_train)
```

This made my model's r2 scores further apart in training vs testing data! This shows that my new function is more overfit than before and implies that I can use a lasso to improve my model:

```python

new_lr.score(X_2_train, y_2_train), new_lr.score(X_2_test, y_2_test)
```

Set up my features in standard scalar so I can fit a new model with an apples to apples comparison for all features and then use lasso. Z_poly_test_submit is a variable that is scalling my new testing data that also has polynomial features:

```python

sc = StandardScaler()

Z_poly_train = sc.fit_transform(X_2_train)
Z_poly_test = sc.transform(X_2_test)
Z_poly_test_submit = sc.transform(X_poly_test)

```

Run a lasso with an automated fit search for a more optimal value for alpha:

```python
# Set up a list of Lasso alphas to check.
#try adjusting tolerance parameter raise from .001
l_alphas = np.logspace(-100, 100, 100)
# Cross-validate over our list of Lasso alphas.
lasso_cv_poly = LassoCV(alphas = l_alphas, cv=5, tol = .01)
# Fit model using best ridge alpha!
lasso_cv_poly.fit(Z_poly_train, y_2_train)
#cite lecture on Lassos and Ridges
```

Now score the train data on my newest fature engineered model:

```python
lasso_cv_poly.score(Z_poly_train, y_2_train), lasso_cv_poly.score(Z_poly_test, y_2_test)
```

Now Generate my predictions for all models so far:

```python
predictions_lasso = lasso_cv.predict(Z_train)
predictions_poly_lasso_test = lasso_cv_poly.predict(Z_poly_test)
predictions_poly_lasso_train = lasso_cv_poly.predict(Z_poly_train)
predictions_lr = lr.predict(X)
predictions_poly = new_lr.predict(X_2_train)
predictions_poly_test = new_lr.predict(X_2_test)
```

Score my model against RMSE and print the results. After playing with RMSE for all models, I am including my final winning model. My final winning model is my feature enginered lasso model with polynomial features (lasso_cv_poly).

```python
rms_final_poly_lasso_model_compared = sqrt(mean_squared_error(y_2_train, predictions_poly_lasso_train)), sqrt(mean_squared_error(y_2_test, predictions_poly_lasso_test))
#rms_poly_test = sqrt(mean_squared_error(y_2_train, predictions_poly_test))
print("RMSE For Feature Enginered Lasso Model: With Polynomial Features (Train VS Test) = " +str(rms_final_poly_lasso_model_compared))
```
